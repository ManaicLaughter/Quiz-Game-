```python
#Starting of Game [Some '--' to just make it look good]
print("__________________________________________________________________________________\n"+
      "----------------------------------------------------------------------------------\n")
print("Welcome To The TactX Quiz\n")
print("----------------------------------------------------------------------------------\n"+
      "----------------------------------------------------------------------------------\n")

#Taking User Input
name = input("May I know your Name: ")
salutation = input("Prefix Mr/Ms/Dr : ")
print("----------------------------------------------------------------------------------\n") 
if salutation.lower() == 'mr':
    print('hello!'.upper() + " " + salutation.upper() + " " + name.upper() + "\nLooking handsome today\n")
if salutation.lower() == 'ms':
    print('hello!'.upper() + " " + salutation.upper() + " " + name.upper() + "\nLooking pretty today\n") 
if salutation.lower() == 'dr':
    print('hello!'.upper() + " " + salutation.upper() + " " + name.upper() + "\nGreat to see you playing, Doctor\n")

print("----------------------------------------------------------------------------------\n")   

#Game Menu
intro = input("Please Select a/b/c\n" + "A) Start Game\n" + "B) Rules\n" + "C) Quit\n")
print(intro)
if intro.lower() == 'a':
    print('Starting Game')
if intro.lower() == 'b':
    print('There are 3 questions in each category\n' + 'After selecting your category. Questions will appear one by one\n' + 
         'For every: \n Correct Answer: +3\n Attempts: 3\n\n Best of Luck!!\n') 
if intro.lower() == 'c':
    print('OK BYE!!\n')
print("----------------------------------------------------------------------------------\n") 

#Initial SCores
score = 0
print(name + " Your Score " + str(score))
print("_________________________________________________________________________________\n")

#Quiz Algorithm for allowing only 3 attempts and giving rewards accordingly
def check_guess(guess, answer):
    global score
    still_guessing = True
    attempt = 0
    while still_guessing and attempt < 3:
        if guess.lower() == answer.lower():
            print('Correct Answer')
            score = score + 3 - attempt
            still_guessing = False
        else:
            if attempt < 2:
                guess = input("Sorry wrong answer. Try again... ")
                attempt = attempt + 1
    if attempt == 3:
        print("Correct answer is " + answer )
        

#Quiz Categories        
cat = input('Starting The Game\n Please Select the Category: a/b/c/d \n A) Space\n B) Chemistry\n C) GK\n D) IT\n')
print(cat)
if cat.upper() == 'A':
    print("CATEGORY: Space")
    guess1 = input('Name the biggest planet of solar system?')
    check_guess(guess1, 'Jupiter')
    guess2 = input('Which is the closest planet to the sun?')
    check_guess(guess2, 'Mercury')
    guess3 = input('Which planet is called Earth\'s Twin?')
    check_guess(guess3, 'Venus')
if cat.upper() == 'B':
    print('CATEGORY: Chemistry')
    guess1 = input('He is the periodic table symbol of which element?')
    check_guess(guess1, 'Helium')
    guess2 = input('Diamond is lustrous because its __________ index is high.')
    check_guess(guess2, 'refractive')
    guess3 = input('Thermometer is an instrument that measures _________.')
    check_guess(guess3, 'Temperature')
if cat.upper() == 'C':
    print('CATEGORY: General Knowledge')
    guess1 = input('Which is the world\'s biggest political party?')
    check_guess(guess1, 'BJP')
    guess2 = input('Which country has highest population?')
    check_guess(guess2, 'China')
    guess3 = input('Which is the largest mammal?')
    check_guess(guess3, 'Blue Whale')
if cat.upper() == 'D':
    print('CATEGORY: Information Technology')
    guess1 = input('Which OS was founded by Bill Gates?')
    check_guess(guess1, 'Microsoft')
    guess2 = input('Name of AI created by IBM?')
    check_guess(guess2, 'Watson')
    guess3 = input('A desktop computer is also known as a ______.')
    check_guess(guess3, 'pc')

#Total Score after completion
print("\n----------------------------------------------------------------------------------\n") 
print( name + ' Your score is ' + str(score))
print("\n----------------------------------------------------------------------------------\n") 

#The End footnote
print("Hope You enjoyed the Quiz\n" + "Keep playing\n\n GOODBYE!!\n" )
print("----------------------------------------------------------------------------------\n" +
     "__________________________________________________________________________________\n")
```
